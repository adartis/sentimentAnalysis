```python
# imports relevant modules matplotlib.pyplot to form graph
import matplotlib.pyplot as plt
# imports numpy
import numpy as np
# imports pandas
import pandas as pd
```

```python
# Load the data (modify the file path as needed)
df = pd.read_csv("news_results_with_text_gpt_processed.csv")
df2 = pd.read_csv("analysis_file_all_functions_applied.csv")
```

```python

# Choose the column to analyse - counts unique values in relevant column
value_counts = df['source_name'].value_counts()

# Convert to DataFrame for manipulation - index created and previous index becomes col. 1.
count_df = value_counts.reset_index()
count_df.columns = ['Source', 'Count']  # Rename columns for df manipulation with matplotlib

# Test to show index 
print(count_df)
```

```python
# Plot as bar chart - source lists identified by title
plt.bar(count_df['Source'], count_df['Count'])

# Labels and title added
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Frequency of Each Value in Column")
plt.xticks(rotation=90)  # Rotate labels for readability

# Show plot
plt.show()
```

```python
# Choose the column to analyse - counts unique values in relevant column - message_title
value_counts = df2['message_title'].value_counts()

# Convert to DataFrame for manipulation - index created and previous index becomes col. 1.
count_df2 = value_counts.reset_index()
count_df2.columns = ['message_title', 'Count']  # Rename columns for df manipulation with matplotlib

# Test to show index 
print(count_df2)
```
